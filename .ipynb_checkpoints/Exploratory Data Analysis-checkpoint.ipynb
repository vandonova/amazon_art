### Cleaning up the size column and filling in misordered widths

```python
print type(amazon_art['size'][0])
print amazon_art['size'][0]

# extract missing height and width information from the size column and cast them as floats
def size_to_width(string):
    dim = string.replace('"','').replace("'",'').replace('u', '').replace('[','').replace(']','').replace('in.','').strip()
    width = float(dim.split(',')[-1].strip())
    return width

def size_to_height(string):
    dim = string.replace('"','').replace("'",'').replace('u', '').replace('[','').replace(']','').replace('in.','').strip()
    height = float(dim.split(',')[-1].strip())
    return height
```

```python
size_to_width("[u'47.24', u'31.5', u'3.94 in.']")
```

```python
amazon_art.loc[amazon_art['width'] < 3, 'width'] = amazon_art.loc[amazon_art['width'] < 3, 'size'].apply(size_to_width)
```

```python
amazon_art.loc[amazon_art['height'] <= 3, 'height'] = amazon_art.loc[amazon_art['height'] <= 3, 'size'].apply(size_to_height)
```

### Create Bins for Price Classification

```python
amazon_art.loc[amazon_art.price_sqin < .50, 'price_class'] = int(0)
amazon_art.loc[(amazon_art.price_sqin >= .50) & (amazon_art.price_sqin < 1), 'price_class'] = int(1)
amazon_art.loc[(amazon_art.price_sqin >= 1) & (amazon_art.price_sqin < 2), 'price_class'] = int(2)
amazon_art.loc[(amazon_art.price_sqin >= 2) & (amazon_art.price_sqin < 10), 'price_class'] = int(3)
amazon_art.loc[(amazon_art.price_sqin >= 10) & (amazon_art.price_sqin < 50), 'price_class'] = int(4)
amazon_art.loc[(amazon_art.price_sqin >= 50) & (amazon_art.price_sqin < 200), 'price_class'] = int(5)
amazon_art.loc[(amazon_art.price_sqin >= 200), 'price_class'] = int(6)
```

# Using Area to Predict Price

Results above imply that maybe price/sqin is not such a good indicator or that size of the print on it's own is not a good predictor.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.cross_validation import train_test_split, cross_val_score
from sklearn import metrics
from nltk import ConfusionMatrix

#trying with more to predict with just area
X = amazon_art[['area']]
y = amazon_art['price_class']

X_train, X_test, y_train, y_test = train_test_split( X, y, random_state=123)

logreg = LogisticRegression()  
logreg.fit(X_train, y_train)
preds = logreg.predict(X_test)

logreg.score(X_train, y_train)
print metrics.accuracy_score(y_test, preds)
print metrics.confusion_matrix(y_test, preds)

y.value_counts()
```

```python
import lda

# Instantiate a count vectorizer with two additional parameters
vect = CountVectorizer(stop_words='english', ngram_range=(1,3)) 
desc = amazon_art.description
sentences_train = vect.fit_transform(desc)

# Instantiate an LDA model
model = lda.LDA(n_topics=6, n_iter=100)
model.fit(sentences_train) # Fit the model 
n_top_words = 10
topic_word = model.topic_word_
for i, topic_dist in enumerate(topic_word):
    topic_words = np.array(vect.get_feature_names())[np.argsort(topic_dist)][:-n_top_words:-1]
    print('Topic {}: {}'.format(i, ', '.join(topic_words)))
```

# Image Feature Extraction from URLs

```python
# converting url into an image file

def url_to_image(url):
    #download the image
    resp = urllib.urlopen(url)
    #convert to NumPy array
    image = np.asarray(bytearray(resp.read()), dtype ='uint8')
    #read it into OpenCV
    image = cv2.imdecode(image, cv2.IMREAD_UNCHANGED)
    return image

def np_array(url):
    image = url_to_image(url)
    cv_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    return cv_rgb

def hsv_array(cv_rgb):
    cv_hsv = cv2.cvtColor(cv_rgb, cv2.COLOR_RGB2HSV)
    return cv_hsv
```

```python
amazon_art['np_img'] = amazon_art.image.apply(np_array)
```

```python
amazon_art['hsv_img'] = amazon_art.np_img.apply(hsv_array)
```

For HSV, OpenCV's Hue range is [0,179], Saturation range is [0,255] and Value range is [0,255]. Different softwares use different scales. So if you are comparing OpenCV values with them, you need to normalize these ranges.

```python
print amazon_art.columns, len(amazon_art.columns)
amazon_art.head(1)
```

```python
# define functions to extract the mean values of red, blue, green, hue, saturation, and total mean value
def avg_rgb_vals(nparray):
    red = nparray[:,:,0].mean()
    green = nparray[:,:,1].mean()
    blue = nparray[:,:,2].mean()
    return red, green, blue

def avg_hsv_vals(nparray):
    hue = nparray[:,:,0].mean()
    saturation = nparray[:,:,1].mean()
    value = nparray[:,:,2].mean()
    return hue, saturation, value
```

```python
# apply .mean() functions to the rgb and hsv values
amazon_art['avg_r_value'], amazon_art['avg_g_value'], amazon_art['avg_b_value'] = zip(*amazon_art.np_img.apply(avg_rgb_vals))
amazon_art['avg_hue'], amazon_art['avg_saturation'], amazon_art['avg_value'] = zip(*amazon_art.hsv_img.apply(avg_hsv_vals))
```

## Using Clustering to Decide on Price Class Splits

```python
amazon_art.columns
```

```python
# K-means with 7 clusters
# define features
feature_cols = ['area', 'avg_r_value', 'avg_g_value', 'avg_b_value', 'avg_hue',
       'avg_saturation', 'avg_value', 'price']

# define X
X = amazon_art[feature_cols]

from sklearn.cluster import KMeans
km = KMeans(n_clusters=7, random_state=1)
km.fit(X)
```

```python
# create a cluster column 
amazon_art['cluster'] = km.labels_
amazon_art.cluster.value_counts()
```

```python
amazon_art.groupby('cluster').mean()
```

```python
# create a "colors" array for plotting
colors = np.array(['red', 'green', 'blue', 'yellow', 'brown', 'purple'])

# scatter plot of area versus price, colored by cluster
plt.scatter(amazon_art.area, amazon_art.price, c=colors[list(amazon_art.cluster)], s=50)

# cluster centers, marked by "+"
#plt.scatter(amazon_art.area, amazon_art.price, linewidths=3, marker='+', s=300, c='black')

# add labels
plt.xlabel('area')
plt.ylabel('price')
```
