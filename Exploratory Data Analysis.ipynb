### Cleaning up the size column and filling in misordered widths

```python
print type(amazon_art['size'][0])
print amazon_art['size'][0]
def size_to_width(string):
    dim = string.replace('"','').replace("'",'').replace('u', '').replace('[','').replace(']','').replace('in.','').strip()
    print dim
    width = float(dim.split(',')[-1].strip())
    return width

def size_to_height(string):
    dim = string.replace('"','').replace("'",'').replace('u', '').replace('[','').replace(']','').replace('in.','').strip()
    height = float(dim.split(',')[-1].strip())
    return height
```

# Image extraction from urls

```python
#urls = []

#for url in amazon_art.image:
#    urls.append(url)
    
#converting url into an image file

def url_to_image(url):
    #download the image
    resp = urllib.urlopen(url)
    #convert to NumPy array
    image = np.asarray(bytearray(resp.read()), dtype ='uint8')
    #read it into OpenCV
    image = cv2.imdecode(image, cv2.IMREAD_UNCHANGED)
    
    return image

#img_array = []
#for url in urls:
#    image = url_to_image(url)
#    cv_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
#    img_array.append(cv_rgb)

def np_array(url):
    image = url_to_image(url)
    cv_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    return cv_rgb
```

```python
art_artist['3D_img'] = amazon_art.image.apply(np_array)
```
